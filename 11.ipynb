```python
Даны две функции:
f(x) = x^3 - 50x

g(x) = -x^4 + 88x^2 - 241
```

```python
from sympy import *
from sympy.abc import x
f = g = Function('f')

f = x ** 3 - 50 * x
g = -x ** 4 + 88 * x ** 2 - 241
```

```python
cross_points = [{'y': f.subs(x, cross_x), 'x': cross_x} for cross_x in solve(f - g, x)]
print('Точек пересечения:', len(cross_points))
```
